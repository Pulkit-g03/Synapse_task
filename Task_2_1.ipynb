```python
frequency= df.groupby('country')['type'].value_counts()
print(frequency)
```

```python
countries = df['country'].str.split(',',expand=True).stack().reset_index(drop=True, level=1)

countries_df = pd.DataFrame({'countriesss': countries, 'type': df['type']})

countries_df
```

```python
frequency= countries_df.groupby('countriesss')['type'].value_counts()
freq= frequency.sort_values(ascending=False)
print(freq)
```
